## Import Data

```python
import pandas as pd
dat = pd.read_csv('epa-sea-level.csv')
dat.head()
```

## Scatter plot of Sea Level

Scatter plot using the Year column as the x-axis and the CSIRO Adjusted Sea Level column as the y-axis.

```python
import matplotlib.pyplot as plt
plt.scatter(dat['Year'], dat['CSIRO Adjusted Sea Level'],color='skyblue')

    
```

From the above plot, we know that the sea level always increase year by year.

## Linregress function

Use the linregress function from scipy.stats to get the slope and y-intercept of the line of best fit.

Plot the line of best fit over the top of the scatter plot. Make the line go through the year 2050 to predict the sea level rise in 2050.

```python
from scipy.stats import linregress
import numpy as np
line_best = linregress(dat['Year'], dat['CSIRO Adjusted Sea Level'])
x_best = np.arange(dat['Year'].min(),2050,1)
y_best = x_best*line_best.slope + line_best.intercept
plt.plot(x_best,y_best,color='red')
```

## Lingress plot from 2000 until 2050

Plot a new line of best fit just using the data from year 2000 through the most recent year in the dataset.

Make the line also go through the year 2050 to predict the sea level rise in 2050 if the rate of rise continues as it has since the year 2000.
The x label should be Year, the y label should be Sea Level (inches), and the title should be Rise in Sea Level.
The boilerplate also includes commands to save and return the image.

```python
df_start = dat[dat['Year'] >= 2000]
line_best2 = linregress(df_start['Year'], df_start['CSIRO Adjusted Sea Level'])
x_best2 = np.arange(2000,2050,1)
y_best2 = x_best2*line_best2.slope + line_best2.intercept
plt.plot(x_best2,y_best2,color='purple')
plt.xlabel('Year')
plt.ylabel('Sea Level (inches)')
plt.title('Rise in Sea Level')
```
